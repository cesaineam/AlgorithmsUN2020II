```
import pydot
from collections import defaultdict 
from IPython.display import Image, display
```

```
def lstr(p):
        return ''.join([str(elem)  for elem in p])
```

```
def fondo(maximo,men,ini,doc):
  maximo.append(men)
  inicial= doc[ini]
  for i in range(len(inicial)):
    
    fondo(maximo,men+1,inicial[i],doc)
    
  return max(maximo)
```

```
def ModifiedBFS(k): 

        # Create dictionaries graph,visited and pydot graph G
        graph = defaultdict()
        visited = defaultdict(int)
        G = pydot.Dot(graph_type="digraph")
        aux=[]
        for i in range(k):
          aux.append(i+1)
        # Create a queue for BFS 
        queue = [] 
  
        # Mark the source node as visited and enqueue it 
        queue.append(aux) 
        visited[lstr(aux)] = 1
        while queue: 


            # Dequeue a vertex from queue 
            s = queue.pop(0) 
            graph[lstr(s)]=[]
            node = pydot.Node(lstr(s))
            G.add_node(node)


            for i in range(len(s)-1):
              d = s[i+1::-1] + s[i+2::]
              if (d != s) and (visited[lstr(d)]==0):                
                node = pydot.Node(lstr(d))
                G.add_node(node)
                edge = pydot.Edge(lstr(s),lstr(d))
                graph[lstr(s)].append(lstr(d))
                G.add_edge(edge)
                queue.append(d)
                visited[lstr(d)]= 1
        aux2 = []
        aux2.append(1)
        
        
        maximo = fondo(aux2,1,lstr(aux),graph)
        print("Pn =",maximo-1)
        # Dsplay the pydot graph G
        im = Image(G.create_png())
        display(im)
```

```
p = 3
ModifiedBFS(p)
```

```
def twoHandFlip(lis, pos1, pos2):
  l1 = []
  l2 = []
  l3 = []
  if pos1 == pos2:
    return lis
  if pos1 > 1:
    l1 = lis[0:pos1-1]
  if pos2 < len(lis):
    l3 = lis[pos2:]
  l2 = lis[pos1-1:pos2]
  l2 = l2[::-1]
  return l1 + l2 + l3
```
