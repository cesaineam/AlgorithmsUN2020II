````
#N numero de hombres, se asume el número de mujeres es igual
def wPrefersM1OverM(prefer, w, m, m1, N): #indica si una mujer prefiere a M1 sobre M2, hombres
	for i in range(N): 
		if (prefer[w][i] == m1): 
			return True
		if (prefer[w][i] == m): 
			return False

def stableMarriage(prefer, N): #Implementa el algoritmo Gale-shapley asumiendo que los hombres se proponen a las mujeres 
	wPartner = [-1 for i in range(N)] 
	mFree = [False for i in range(N)] 

	freeCount = N 
	while (freeCount > 0): 
		m = 0
		while (m < N): 
			if (mFree[m] == False): 
				break
			m += 1
		i = 0
		while i < N and mFree[m] == False: 
			w = prefer[m][i] 
			if (wPartner[w - N] == -1): 
				wPartner[w - N] = m 
				mFree[m] = True
				freeCount -= 1

			else: 
				m1 = wPartner[w - N] 
				if (wPrefersM1OverM(prefer, w, m, m1, N) == False): 
					wPartner[w - N] = m 
					mFree[m] = True
					mFree[m1] = False
			i += 1
	print("Mujer ", " Hombre") 
	for i in range(N): 
		print(i + N, "\t", wPartner[i])
````

````
#N numero de hombres, se asume el número de mujeres es igual
def wPrefersM1OverM(prefer, w, m, m1, N): #indica si una mujer prefiere a M1 sobre M2, hombres
	for i in range(N): 
		if (prefer[w][i] == m1): 
			return True
		if (prefer[w][i] == m): 
			return False

def stableMarriage(prefer, N): #Implementa el algoritmo Gale-shapley asumiendo que los hombres se proponen a las mujeres 
  wPartner = [-1 for i in range(N)]
  mFree = [False for i in range(N)] 
  it = 0
  freeCount = N 
  while (freeCount > 0): 
    m = 0
    while (m < N): 
      if (mFree[m] == False): 
        break
      m += 1
    i = 0
    while i < N and mFree[m] == False: 
      it += 1
      w = prefer[m][i] 
      if (wPartner[w - N] == -1): 
        wPartner[w - N] = m 
        mFree[m] = True
        freeCount -= 1
      else: 
        m1 = wPartner[w - N] 
        if (wPrefersM1OverM(prefer, w, m, m1, N) == False): 
          wPartner[w - N] = m 
          mFree[m] = True
          mFree[m1] = False
        i += 1
  return it
````
